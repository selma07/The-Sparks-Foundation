### Data Visualisation

```python
viz = df_sample[['Ship Mode', 'Segment', 'Country', 'City', 'State', 'Postal Code','Region', 'Category', 'Sub-Category', 'Sales', 'Quantity', 'Discount', 'Profit']]
viz.hist()

plt.show()
```

```python
sns.countplot(df_sample['State'])
```

```python
sns.countplot(df_sample['Segment'])
```

```python
sns.countplot(df_sample['Region'])
```

```python
sns.countplot(df_sample['Region'],order=(df_sample['Region'].value_counts().head(30)).index)
plt.xticks(rotation=90)
```

```python
df_sample.Segment.value_counts().plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%', # add in percentages    
                            shadow=True,       # add shadow      
                            )

plt.title('Sample Segment')
plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```

```python
df_sample.Region.value_counts().plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%', # add in percentages    
                            shadow=True,       # add shadow      
                            )

plt.title('Sample Region')
plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```

```python
labels = df_sample["Segment"].unique()
data = df_sample.groupby("Segment").sum()
profit = data["Profit"]

plt.pie(profit, labels=labels, shadow=True, autopct="%1.1f%%", startangle=90)
plt.title("Segments and their total profit")
plt.show()
```

```python
data = df_sample.loc[:, ['Region', 'Profit']]

data = data.groupby(['Region'],as_index=False).sum()

sns.barplot(x='Profit',y='Region',data=data).set_title('Profit in Region')
sns.set_style('whitegrid')
```

```python
data = df_sample.loc[:, ['Segment', 'Profit']]

data = data.groupby(['Segment'],as_index=False).sum()

sns.barplot(x='Profit',y='Segment',data=data).set_title('Profit in Segment')
sns.set_style('whitegrid')
```

```python
data = df_sample.loc[:, ['Segment', 'Discount']]

data = data.groupby(['Segment'],as_index=False).sum()

sns.barplot(x='Discount',y='Segment',data=data).set_title('Discount in Segment ')
sns.set_style('whitegrid')
```

```python
pd.DataFrame(df_sample.groupby('State').sum())['Profit'].sort_values(ascending=True)
```

```python
plt.figure(figsize = (10,7))
sns.lineplot('Region', 'Profit', data = df_sample, color = 'g', label= 'Region')
plt.legend()
```

```python
plt.figure(figsize = (10, 7))
sns.lineplot('Region', 'Discount', data = df_sample, color = 'g', label= 'Region')
plt.legend()
```
